## Step 4 - List Available Buckets

The gsutil command is how we will access GCP's storage for our project from our terminal. Let's take a look at the bucket we set up earlier.


Run the command below:

```bash
$  gsutil ls
```

By doing so we can see all of the listed storage buckets associated with the current GCP project.

Now let's use the full path to look at the objects inside of the bucket.

```bash
gsutil ls gs://[bucketpath]
```
<img src="../assets/images/Data-in_the_Cloud_13.png" style="max-width:100%;border-width:3px;  border-style: solid;" />

## Step 5 - Moving files to the GCP using the local terminal

Now we will use gsutil to put the flight data csvs into the bucket.

Let's copy these files from the local directory to the GCP bucket. You will need the file path of the bucket again.

```bash
$  gsutil cp [file-name] gs://[bucketpath]

$  gsutil cp  *.csv   gs://[bucketpath]/flights-data/

```



```bash
gsutil ls gs://[bucketpath]/flights-data
```

The output might look as below (exact output may differ)

```
gs://danielson_bucket_1/flights-data/201501.csv
gs://danielson_bucket_1/flights-data/201502.csv
gs://danielson_bucket_1/flights-data/201503.csv
```

And we can see that the files are now in our bucket.
